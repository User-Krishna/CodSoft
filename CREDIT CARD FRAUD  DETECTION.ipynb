```python
df_train=pd.read_csv("credit/fraudTrain.csv")
df_test=pd.read_csv("credit/fraudTest.csv")
print(df_train.head())
df_test.head()
```

```python
df_train.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
df_test.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
```

```python
df_train.drop(columns=['trans_date_trans_time','first', 'last','job', 'dob', 'street', 'city', 'state', 'zip','trans_date_trans_time','merchant','category', ],inplace=True)
df_test.drop(columns=['trans_date_trans_time','first', 'last','job', 'dob', 'street', 'city', 'state', 'zip','trans_date_trans_time','merchant','category', ],inplace=True)
```

```python
dummies_train=pd.get_dummies(df_train['gender'])
dummies_test=pd.get_dummies(df_test['gender'])
print(dummies_train)
dummies_test
```

```python
df_train
```

```python
df_train=pd.concat([df_train,dummies_train],axis=1)
df_test=pd.concat([df_test,dummies_test],axis=1)
```

```python
df_train
```

```python
df_test
```

```python
y_train=df_train['is_fraud']
y_test=df_test['is_fraud']
print(y_train)
y_test
```

```python
df_train.drop(columns=['is_fraud','gender'],inplace=True)
df_test.drop(columns=['is_fraud','gender'],inplace=True)
```

```python
x_train=df_train
x_test=df_test
print(x_train)
print(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

x_train['trans_num_encoded'] = label_encoder.fit_transform(x_train['trans_num'])
x_test['trans_num_encoded'] = label_encoder.fit_transform(x_test['trans_num'])
x_train.drop('trans_num', axis=1, inplace=True)
x_test.drop('trans_num', axis=1, inplace=True)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python

from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))

report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```
